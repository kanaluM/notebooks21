```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def exp(a,b,c,d):
    """Exponential functions of the form y = a b^(cx) + d"""
    
    x = np.linspace(-5,5,50)
    y = np.asarray(a*b**(c*x)+d)
    fig,ax = plt.subplots(figsize=(5,5))
    ax.plot(x,y)
    ax.set(title=f"Exponential Function: y = ({a}){b}^({c}x)+{d}")
    ax.grid()
```

```python
exp(1,2.71,1,0)
```
